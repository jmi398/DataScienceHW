```python
# Import Python SQL toolkit and Object Relational Mapper
import sqlalchemy
from sqlalchemy.ext.automap import automap_base
from sqlalchemy.orm import Session
from sqlalchemy import create_engine, inspect, func, desc
import datetime as dt
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
# Choose a start date and end date for your trip of 3-15 days total
# start_date = '2010-01-01'
# end_date = '2010-01-04'
```

```python
# Use create_engine to connect to your sqlite database
engine = create_engine("sqlite:///hawaii.sqlite", echo=False)
```

```python
# Declare a Base using `automap_base()`
Base = automap_base()
```

```python
# Use the Base class to reflect the database tables
Base.prepare(engine, reflect=True)
```

```python
# Print all of the classes mapped to the Base
Base.classes.keys()
```

```python
# Save a reference to those classes called Station and Measurement
Measurement = Base.classes.measurement
Station = Base.classes.station
```

```python
# Precipitation Analysis
```

```python
# Create a session
session = Session(engine)
```

```python
# date 1 year ago from today
year_ago = dt.date.today() - dt.timedelta(days=365)
today = dt.date.today()
```

```python
# Design a query to retrieve the last 12 months of precipitation data
precipitation_data = session.query(Measurement.date, Measurement.prcp).\
    filter(Measurement.date > year_ago).\
    filter (Measurement.date < today).all()
precipitation_data    
```

```python
# Plot the results using the DataFrame plot method
df = pd.DataFrame(precipitation_data, columns=['Date', 'Precipitation'])
df.set_index('Date', inplace=True)
df.plot(x_compat=True)
plt.xticks(rotation='45')
plt.show()
```

```python
# Print the summary statistics for the precipitation data
df.describe()
```

```python
# Station Analysis
```

```python
# A query to calculate the total number of stations
total_stations = session.query(func.count(Station.station)).all()
print("Total stations:", total_stations)
```

```python
# Design a query to find the most active stations
# List the stations and observation counts in descending order
active_stations_data = session.query(Measurement.station, func.count(Measurement.tobs)).\
                       group_by(Measurement.station).\
                       order_by(desc(func.count(Measurement.tobs))).all()

active_stations_data   

# Station USC00519281 has the highest number of observations
```

```python
# A query to retrieve the last 12 months of temperature observation data (tobs)
temperature_observation_data = session.query(Measurement.date ,Measurement.tobs).\
    filter(Measurement.date > year_ago).\
    filter (Measurement.date < today).\
    filter (Measurement.station == 'USC00519281').\
    group_by(Measurement.tobs).\
    order_by(Measurement.tobs.desc()).all()   
    
temperature_observation_data    
```

```python
frequency_data = session.query(Measurement.tobs).\
                       filter (Measurement.station == 'USC00519281').all()

# # frequency_data = [ i[0] for i in frequency_data ]
```

```python
# Calc_temps that will accept a start date and end date in the format %Y-%m-%d 
# and returns the minimum, average, and maximum temperatures for that range of dates
def calc_temps ():
        start_datestr = input('Please enter a start date using this format: %Y-%m-%d')
        start_date = dt.datetime.strptime(start_datestr, '%Y-%m-%d')
        end_datestr = input('Please enter an end date using this format: %Y-%m-%d')
        end_date = dt.datetime.strptime(end_datestr,'%Y-%m-%d')
        minimum_temp = session.query(func.min(Measurement.tobs)).\
                    filter(Measurement.date >= start_date).\
                    filter(Measurement.date <= end_date).all()
        maximum_temp = session.query(func.max(Measurement.tobs)).\
                    filter(Measurement.date >= start_date).\
                    filter(Measurement.date <= end_date).all()
        average_temp = session.query(func.avg(Measurement.tobs)).\
                    filter(Measurement.date >= start_date).\
                    filter(Measurement.date <= end_date).all()       
        
        return("The minimum temperature is:", minimum_temp, 
               ", the maximum temperature is:", maximum_temp,
              ", and the average temperature is", average_temp)
```
